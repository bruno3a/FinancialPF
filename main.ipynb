```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
blue = pd.read_csv('H:\My Drive\Formacion\Data Science\Financial\FinancialPF\CSVs\Blue.csv')
CER = pd.read_csv('H:\My Drive\Formacion\Data Science\Financial\FinancialPF\CSVs\CER.csv')
MEP = pd.read_csv('H:\My Drive\Formacion\Data Science\Financial\FinancialPF\CSVs\MEP.csv')
TaxedD = pd.read_csv('H:\My Drive\Formacion\Data Science\Financial\FinancialPF\CSVs\TaxedDollar.csv')
```

```python
df = blue.merge(MEP, on='Date', how= 'left', suffixes=['blue'], sort=True)
df = df.merge(CER, on='Date', how= 'left', suffixes=['Value_CER'], sort=True) # If we had updated CER values, we'd use Right method, to keep latest values of CER
df = df.merge(TaxedD, on='Date', how= 'left', suffixes=['_MEP','_Taxed'], sort=True)
df.rename(columns={'Buy':'blue_Buy', 'Sell':'blue_Sell'}, inplace=True)
```

```python
df.Date = pd.to_datetime(df.Date, format='%d-%m-%Y')
```
